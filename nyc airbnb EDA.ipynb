```python
# changing type to string and cleaning
df['name'] = df['name'].astype(str)
def clean_text(text):
    cleaned_text = ''.join(e for e in text if e.isalnum() or e.isspace())
    return ' '.join(cleaned_text.split())
df['name'] = df['name'].apply(clean_text)
df
```

```python
# counting listings in neighbourhood groups
sns.set(rc = {'figure.figsize':(10,5)})
ax = sns.countplot(data = df , x = 'neighbourhood_group')
ax
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['neighbourhood_group'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False)
```

```python
# neighbourhood group and price comparison
neighbourhood_price = df.groupby(['neighbourhood_group'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False)
sns.barplot(x = 'neighbourhood_group',y = 'price',data =neighbourhood_price )
```

```python
# room type count
sns.set(rc = {'figure.figsize':(10,5)})
ax = sns.countplot(data = df , x = 'room_type')
ax
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['room_type'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False)
```

```python
# room type and price comparison
neighbourhood_type = df.groupby(['room_type'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False)
sns.barplot(x = 'room_type',y = 'price',data =neighbourhood_type )
```

```python
df.groupby(['neighbourhood'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False).head(10)
```

```python
# prices across different neighbourhoods
neighbour_price = df.groupby(['neighbourhood'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False).head(10)
sns.barplot(x = 'neighbourhood',y = 'price',data =neighbour_price )
sns.set(rc = {'figure.figsize':(20,5)})
```

```python
df.groupby(['host_name'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False).head(10)
```

```python
# hosts with the highest prices
host_price = df.groupby(['host_name'],as_index = False)['price'].sum().sort_values(by = 'price',ascending = False).head(10)
sns.barplot(x = 'host_name',y = 'price',data =host_price )
sns.set(rc = {'figure.figsize':(20,5)})
```

```python
df.groupby(['host_name'],as_index = False)['number_of_reviews'].sum().sort_values(by = 'number_of_reviews',ascending = False).head(10)
```

```python
# hosts with the highest reviews
host_reviews = df.groupby(['host_name'],as_index = False)['number_of_reviews'].sum().sort_values(by = 'number_of_reviews',ascending = False).head(10)
sns.barplot(x = 'host_name',y = 'number_of_reviews',data =host_reviews )
```
